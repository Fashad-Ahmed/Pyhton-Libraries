```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1, 20, 50)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
# plt.style.use('')
plt.show()
```

```python
x = np.linspace(1, 20, 50)
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '-.')
plt.style.use('default')
# plt.show()
```

```python
# x = np.linspace(1, 20, 50)
fig = plt.figure()

plt.subplot(2,1,1)
plt.plot(x, np.sin(x), '-')

plt.subplot(2,1,2)
plt.plot(x, np.cos(x), '--')
```
