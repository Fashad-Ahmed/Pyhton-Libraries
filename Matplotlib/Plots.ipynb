```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(5,10,3)
x
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,2,2])

ax.plot(x, x**2, label="SQUARED")
ax.plot(x, x**3, label="CUBED")
ax.plot(x, x**4, label="QUADRAPLED")

ax.legend(loc=(0.2, 0.9))
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x**4, x**8, color="purple", lw=5, ls='--')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x**24, x**6, color="purple", lw=2, ls='-', marker="o", markersize=20, markerfacecolor="blue", markeredgewidth=3, markeredgecolor="yellow" )
```
